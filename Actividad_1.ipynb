## Primer ejercicio
A la mamá de Juan le preguntan su edad, y contesta: tengo 3 hijos. pregúntele a Juan su edad. Alberto tiene 2/3 de la edad de Juan. Ana tiene 4/3 de la edad de Juan y mi edad es la suma de las tres. Hacer un algoritmo que muestre la edad de los cuatro.

````python
class Persona:
    def __init__(self, nombre, edad=0.0):
        self.nombre = nombre
        self.edad = float(edad)

    def mostrar_edad(self):
        # Se aproxima al entero mas cercano
        print(f"La edad de {self.nombre} es: {self.edad:.0f} años")


class Familia:
    def __init__(self, edad_juan):
        edad_juan = float(edad_juan)

        self.juan = Persona("Juan", edad_juan)
        self.alberto = Persona("Alberto", (2/3) * edad_juan)
        self.ana = Persona("Ana", (4/3) * edad_juan)
        self.mama = Persona("Mamá",
                            self.juan.edad +
                            self.alberto.edad +
                            self.ana.edad)

    def mostrar_edades(self):
        print("\nEdades de la familia:\n")
        self.juan.mostrar_edad()
        self.alberto.mostrar_edad()
        self.ana.mostrar_edad()
        self.mama.mostrar_edad()



if __name__ == "__main__":
    edad_juan = float(input("Ingrese la edad de Juan: "))
    familia = Familia(edad_juan)
    familia.mostrar_edades()
````

## Ejercicio 3

Un empleado trabaja 48 horas en la semana a razón de $5000 hora. El porcentaje de retención en la fuente es del 12,5% del salario bruto. Se desea saber cuál es el salario bruto. la retención en la fuente y el salario neto del trabajador.

````python
class Empleado:
    def __init__(self, horas_trabajadas, valor_hora, porcentaje_retencion):
        self.horas = horas_trabajadas
        self.valor_hora = valor_hora
        self.porcentaje_retencion = porcentaje_retencion / 100

    def salario_bruto(self):
        return self.horas * self.valor_hora

    def retencion(self):
        return self.salario_bruto() * self.porcentaje_retencion

    def salario_neto(self):
        return self.salario_bruto() - self.retencion()

    def mostrar_resultados(self):
        print(f"Salario Bruto: ${self.salario_bruto():.0f}")
        print(f"Retención en la fuente: ${self.retencion():.0f}")
        print(f"Salario Neto: ${self.salario_neto():.0f}")


empleado = Empleado(48, 5000, 12.5)
empleado.mostrar_resultados()
````

## Ejercicio 4

Elabore un algoritmo que lea un número y obtenga su cuadrado y su cubo.

````python
import math

class Numero:
    def __init__(self, valor):
        self.valor = float(valor)

    def cuadrado(self):
        return math.pow(self.valor, 2)

    def cubo(self):
        return math.pow(self.valor, 3)

    def mostrar_resultados(self):
        print("Número:", self.valor)
        print("Cuadrado:", self.cuadrado())
        print("Cubo:", self.cubo())


num = float(input("Ingrese un número: "))
obj = Numero(num)
obj.mostrar_resultados()
````

## Ejercicio 5

Dado el radio de un círculo. Haga un algoritmo que obtenga el área del círculo y la longitud de la circunferencia.

````python
import math

class Circulo:
    def __init__(self, radio):
        self.radio = float(radio)

    def calcular_area(self):
        return math.pi * math.pow(self.radio, 2)

    def calcular_circunferencia(self):
        return 2 * math.pi * self.radio

    def mostrar_resultados(self):
        print(f"Radio: {self.radio}")
        print(f"Área: {self.calcular_area():.2f}")
        print(f"Longitud de la circunferencia: {self.calcular_circunferencia():.2f}")


radio = float(input("Ingrese el radio del círculo: "))
circulo = Circulo(radio)
circulo.mostrar_resultados()
````
