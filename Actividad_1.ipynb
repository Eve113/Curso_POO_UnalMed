## Primer ejercicio
A la mamá de Juan le preguntan su edad, y contesta: tengo 3 hijos. pregúntele a Juan su edad. Alberto tiene 2/3 de la edad de Juan. Ana tiene 4/3 de la edad de Juan y mi edad es la suma de las tres. Hacer un algoritmo que muestre la edad de los cuatro.

```python
class Persona:
    def __init__(self, nombre, edad=0.0):
        self.nombre = nombre
        self.edad = float(edad)

    def mostrar_edad(self):
        # Se aproxima al entero mas cercano
        print(f"La edad de {self.nombre} es: {self.edad:.0f} años")


class Familia:
    def __init__(self, edad_juan):
        edad_juan = float(edad_juan)

        self.juan = Persona("Juan", edad_juan)
        self.alberto = Persona("Alberto", (2/3) * edad_juan)
        self.ana = Persona("Ana", (4/3) * edad_juan)
        self.mama = Persona("Mamá",
                            self.juan.edad +
                            self.alberto.edad +
                            self.ana.edad)

    def mostrar_edades(self):
        print("\nEdades de la familia:\n")
        self.juan.mostrar_edad()
        self.alberto.mostrar_edad()
        self.ana.mostrar_edad()
        self.mama.mostrar_edad()



if __name__ == "__main__":
    edad_juan = float(input("Ingrese la edad de Juan: "))
    familia = Familia(edad_juan)
    familia.mostrar_edades()
```
