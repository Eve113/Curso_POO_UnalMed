# <font color=#6130b0 size=6>**Ejercicio 8.2. Definicion de clases**</font>

Se requiere desarrollar un programa con interfaz gráfica de usuario que genere una ventana donde se solicite el ingreso de cinco notas obtenidas por un estudiante. El programa debe calcular y mostrar en la parte inferior de la ventana los siguientes datos:



*   El promedio de notas ingresadas.
*   La desviación estándar de las notas ingresadas.
*   La mayor nota obtenida.
*   La menor nota obtenida.




```python
# CLASE NOTAS
class Notas:

    def __init__(self, notas):

        self.notas = notas

    # CALCULOS
    def calcular_promedio(self):

        return sum(self.notas) / len(self.notas)

    def calcular_desviacion(self):

        return statistics.stdev(self.notas)

    def calcular_mayor(self):

        return max(self.notas)

    def calcular_menor(self):

        return min(self.notas)


# CLASE VENTANA PRINCIPAL
class VentanaPrincipal:

    def __init__(self):

        self.titulo = widgets.HTML(
            value="""
            <h1 style='color:#6130b0; text-align:center;'>
                Sistema de Gestión de Notas
            </h1>

            <h3 style='text-align:center; color:gray;'>
                Ingrese sus respectivas notas para obtener los resultados correspondientes.
            </h3>
            """
        )

        estilo = {'description_width': '80px'}

        layout_caja = widgets.Layout(
            width='250px',
            margin='10px auto'
        )

        self.nota1 = widgets.FloatText(
            description='Nota 1:',
            style=estilo,
            layout=layout_caja
        )

        self.nota2 = widgets.FloatText(
            description='Nota 2:',
            style=estilo,
            layout=layout_caja
        )

        self.nota3 = widgets.FloatText(
            description='Nota 3:',
            style=estilo,
            layout=layout_caja
        )

        self.nota4 = widgets.FloatText(
            description='Nota 4:',
            style=estilo,
            layout=layout_caja
        )

        self.nota5 = widgets.FloatText(
            description='Nota 5:',
            style=estilo,
            layout=layout_caja
        )

        self.boton_calcular = widgets.Button(
            description='Calcular Resultados',
            button_style='success',
            icon='check',
            layout=widgets.Layout(
                width='250px',
                height='45px',
                margin='20px auto'
            )
        )

        self.boton_borrar = widgets.Button(
            description='Borrar',
            button_style='danger',
            icon='trash',
            layout=widgets.Layout(
                width='250px',
                height='45px',
                margin='10px auto'
            )
        )

        self.salida = widgets.Output(
            layout=widgets.Layout(
                display='flex',
                justify_content='center'
            )
        )

        # EVENTOS
        self.boton_calcular.on_click(self.calcular_resultados)
        self.boton_borrar.on_click(self.borrar_notas)

    def obtener_notas(self):

        notas = [
            self.nota1.value,
            self.nota2.value,
            self.nota3.value,
            self.nota4.value,
            self.nota5.value
        ]

        return notas

    # VALIDACIÓN DE NOTAS
    def validar_notas(self, notas):

        for nota in notas:

            if nota < 0 or nota > 5:
                return False

        return True

    # CALCULO DE RESULTADOS
    def calcular_resultados(self, b):

        with self.salida:

            self.salida.clear_output()

            notas = self.obtener_notas()

            # VALIDACIÓN
            if not self.validar_notas(notas):

                display(HTML("""
                <h3 style='color:red; text-align:center;'>
                    Error: Las notas deben estar entre 0 y 5.
                </h3>
                """))

                return

            sistema_notas = Notas(notas)

            # CÁLCULOS
            promedio = sistema_notas.calcular_promedio()
            desviacion = sistema_notas.calcular_desviacion()
            mayor = sistema_notas.calcular_mayor()
            menor = sistema_notas.calcular_menor()

            # RESULTADOS
            display(HTML(f"""
            <div style="
                background-color:#F5F5F5;
                padding:25px;
                border-radius:15px;
                border:2px solid #6130b0;
                width:450px;
                margin:auto;
                text-align:left;
                color:#6b6b6b;
            ">

            <h2 style='color:#6130b0; text-align:center;'>
                Resultados
            </h2>

            <p style='font-size:18px;'>
                📘 <b>Promedio:</b> {promedio:.2f}
            </p>

            <p style='font-size:18px;'>
                📊 <b>Desviación estándar:</b> {desviacion:.2f}
            </p>

            <p style='font-size:18px;'>
                🟢 <b>Mayor nota:</b> {mayor:.2f}
            </p>

            <p style='font-size:18px;'>
                🔴 <b>Menor nota:</b> {menor:.2f}
            </p>

            </div>
            """))

    # BORRAR NOTAS
    def borrar_notas(self, b):

        self.nota1.value = 0
        self.nota2.value = 0
        self.nota3.value = 0
        self.nota4.value = 0
        self.nota5.value = 0

        self.salida.clear_output()

    def mostrar_interfaz(self):

        formulario = widgets.VBox(

            [
                self.titulo,
                self.nota1,
                self.nota2,
                self.nota3,
                self.nota4,
                self.nota5,
                self.boton_calcular,
                self.boton_borrar,
                self.salida
            ],

            layout=widgets.Layout(
                align_items='center'
            )
        )

        display(formulario)


# CLASE PRINCIPAL
class Principal:

    @staticmethod
    def main():

        ventana = VentanaPrincipal()

        ventana.mostrar_interfaz()


# EJECUTAR PROGRAMA
Principal.main()
```

# <font color=#6130b0 size=6>**Ejercicio 8.3. Gestión de eventos**</font>
Se requiere desarrollar un programa con interfaz gráfica de usuario que permita calcular el volumen y superficie de varias figuras geométricas. Las figuras geométricas son el cilindro, la esfera y la pirámide.


*   Para el cilindro se solicitan su radio y altura (en centímetros).
*   Para la esfera, su radio (en centímetros).
*    Para la pirámide, su base, altura y apotema (en centímetros).


Una vez ingresados estos datos, el programa calcula el volumen y superficie de cada figura. Para desarrollar el programa se debe crear una jerarquía de clases para las diferentes figuras geométricas requeridas.

```python

class FiguraGeometrica:

    def __init__(self):

        self.__volumen = 0
        self.__superficie = 0

    def get_volumen(self):
        return self.__volumen

    def set_volumen(self, volumen):
        self.__volumen = volumen

    def get_superficie(self):
        return self.__superficie

    def set_superficie(self, superficie):
        self.__superficie = superficie


# SUBCLASE CILINDRO

class Cilindro(FiguraGeometrica):

    def __init__(self, radio, altura):

        super().__init__()

        self.__radio = radio
        self.__altura = altura

    def calcular_volumen(self):

        volumen = math.pi * (self.__radio ** 2) * self.__altura

        self.set_volumen(volumen)

        return volumen

    def calcular_superficie(self):

        superficie = 2 * math.pi * self.__radio * (
            self.__radio + self.__altura
        )

        self.set_superficie(superficie)

        return superficie



# SUBCLASE ESFERA

class Esfera(FiguraGeometrica):

    def __init__(self, radio):

        super().__init__()

        self.__radio = radio

    def calcular_volumen(self):

        volumen = (4/3) * math.pi * (self.__radio ** 3)

        self.set_volumen(volumen)

        return volumen

    def calcular_superficie(self):

        superficie = 4 * math.pi * (self.__radio ** 2)

        self.set_superficie(superficie)

        return superficie


# SUBCLASE PIRÁMIDE


class Piramide(FiguraGeometrica):

    def __init__(self, base, altura, apotema):

        super().__init__()

        self.__base = base
        self.__altura = altura
        self.__apotema = apotema

    def calcular_volumen(self):

        volumen = (self.__base ** 2 * self.__altura) / 3

        self.set_volumen(volumen)

        return volumen

    def calcular_superficie(self):

        area_base = self.__base ** 2

        area_lateral = 2 * self.__base * self.__apotema

        superficie = area_base + area_lateral

        self.set_superficie(superficie)

        return superficie


# CLASE VentanaCilindro

class VentanaCilindro:

    def __init__(self):

        self.label_titulo = widgets.HTML("""
        <h2 style='color:#6130b0;'>Cilindro</h2>
        """)

        self.txt_radio = widgets.FloatText(
            description='Radio:',
            layout=widgets.Layout(width='300px')
        )

        self.txt_altura = widgets.FloatText(
            description='Altura:',
            layout=widgets.Layout(width='300px')
        )

        self.boton_calcular = widgets.Button(
            description='Calcular Cilindro',
            button_style='success',
            icon='calculator',
            layout=widgets.Layout(width='300px')
        )

        self.resultado = widgets.Output()

        self.boton_calcular.on_click(self.actionPerformed)


    def actionPerformed(self, b):

        with self.resultado:

            self.resultado.clear_output()

            cilindro = Cilindro(
                self.txt_radio.value,
                self.txt_altura.value
            )

            volumen = cilindro.calcular_volumen()

            superficie = cilindro.calcular_superficie()

            display(HTML(f"""
            <div style="
                background:#F5F5F5;
                padding:20px;
                border-radius:15px;
                border:2px solid #6130b0;
                color:#6b6b6b;
                margin-top:15px;
            ">

            <h3 style='color:#6130b0;'>
                Resultados Cilindro
            </h3>

            <p><b>Volumen:</b> {volumen:.2f} cm³</p>

            <p><b>Superficie:</b> {superficie:.2f} cm²</p>

            </div>
            """))

    # INICIO DE LA VENTANA

    def inicio(self):

        return widgets.VBox(
            [
                self.label_titulo,
                self.txt_radio,
                self.txt_altura,
                self.boton_calcular,
                self.resultado
            ],

            layout=widgets.Layout(
                align_items='center'
            )
        )


# CLASE VentanaEsfera

class VentanaEsfera:

    def __init__(self):

        self.label_titulo = widgets.HTML("""
        <h2 style='color:#6130b0;'>Esfera</h2>
        """)

        self.txt_radio = widgets.FloatText(
            description='Radio:',
            layout=widgets.Layout(width='300px')
        )

        self.boton_calcular = widgets.Button(
            description='Calcular Esfera',
            button_style='success',
            icon='calculator',
            layout=widgets.Layout(width='300px')
        )

        self.resultado = widgets.Output()

        self.boton_calcular.on_click(self.actionPerformed)

    def actionPerformed(self, b):

        with self.resultado:

            self.resultado.clear_output()

            esfera = Esfera(self.txt_radio.value)

            volumen = esfera.calcular_volumen()

            superficie = esfera.calcular_superficie()

            display(HTML(f"""
            <div style="
                background:#F5F5F5;
                padding:20px;
                border-radius:15px;
                border:2px solid #6130b0;
                color:#6b6b6b;
                margin-top:15px;
            ">

            <h3 style='color:#6130b0;'>
                Resultados Esfera
            </h3>

            <p><b>Volumen:</b> {volumen:.2f} cm³</p>

            <p><b>Superficie:</b> {superficie:.2f} cm²</p>

            </div>
            """))

    def inicio(self):

        return widgets.VBox(
            [
                self.label_titulo,
                self.txt_radio,
                self.boton_calcular,
                self.resultado
            ],

            layout=widgets.Layout(
                align_items='center'
            )
        )


# CLASE VentanaPiramide

class VentanaPiramide:

    def __init__(self):

        self.label_titulo = widgets.HTML("""
        <h2 style='color:#6130b0;'>Pirámide</h2>
        """)

        self.txt_base = widgets.FloatText(
            description='Base:',
            layout=widgets.Layout(width='300px')
        )

        self.txt_altura = widgets.FloatText(
            description='Altura:',
            layout=widgets.Layout(width='300px')
        )

        self.txt_apotema = widgets.FloatText(
            description='Apotema:',
            layout=widgets.Layout(width='300px')
        )

        self.boton_calcular = widgets.Button(
            description='Calcular Pirámide',
            button_style='success',
            icon='calculator',
            layout=widgets.Layout(width='300px')
        )

        self.resultado = widgets.Output()

        self.boton_calcular.on_click(self.actionPerformed)

    def actionPerformed(self, b):

        with self.resultado:

            self.resultado.clear_output()

            piramide = Piramide(
                self.txt_base.value,
                self.txt_altura.value,
                self.txt_apotema.value
            )

            volumen = piramide.calcular_volumen()

            superficie = piramide.calcular_superficie()

            display(HTML(f"""
            <div style="
                background:#F5F5F5;
                padding:20px;
                border-radius:15px;
                border:2px solid #6130b0;
                color:#6b6b6b;
                margin-top:15px;
            ">

            <h3 style='color:#6130b0;'>
                Resultados Pirámide
            </h3>

            <p><b>Volumen:</b> {volumen:.2f} cm³</p>

            <p><b>Superficie:</b> {superficie:.2f} cm²</p>

            </div>
            """))

    def inicio(self):

        return widgets.VBox(
            [
                self.label_titulo,
                self.txt_base,
                self.txt_altura,
                self.txt_apotema,
                self.boton_calcular,
                self.resultado
            ],

            layout=widgets.Layout(
                align_items='center'
            )
        )


# CLASE VentanaPrincipal

class VentanaPrincipal:

    def __init__(self):

        self.boton_cilindro = widgets.Button(
            description='Abrir Cilindro',
            button_style='primary',
            layout=widgets.Layout(width='250px')
        )

        self.boton_esfera = widgets.Button(
            description='Abrir Esfera',
            button_style='primary',
            layout=widgets.Layout(width='250px')
        )

        self.boton_piramide = widgets.Button(
            description='Abrir Pirámide',
            button_style='primary',
            layout=widgets.Layout(width='250px')
        )

        self.boton_piramide.style.button_color = '#1e9ab3'

        self.boton_esfera.style.button_color = '#1e9ab3'

        self.boton_cilindro.style.button_color = '#1e9ab3'

        self.ventana_cilindro = VentanaCilindro()

        self.ventana_esfera = VentanaEsfera()

        self.ventana_piramide = VentanaPiramide()

        self.boton_cilindro.on_click(self.mostrar_cilindro)

        self.boton_esfera.on_click(self.mostrar_esfera)

        self.boton_piramide.on_click(self.mostrar_piramide)

        self.salida = widgets.Output()



    def mostrar_cilindro(self, b):

        with self.salida:

            self.salida.clear_output()

            display(self.ventana_cilindro.inicio())

    def mostrar_esfera(self, b):

        with self.salida:

            self.salida.clear_output()

            display(self.ventana_esfera.inicio())

    def mostrar_piramide(self, b):

        with self.salida:

            self.salida.clear_output()

            display(self.ventana_piramide.inicio())


    def inicio(self):

        titulo = widgets.HTML("""
        <h1 style='color:#6130b0; text-align:center;'>
            Sistema de Figuras Geométricas
        </h1>

         <h3 style='text-align:center; color:gray;'>
            Escoja la figura a calcular
        </h3>

        """)

        interfaz = widgets.VBox(

            [
                titulo,
                self.boton_cilindro,
                self.boton_esfera,
                self.boton_piramide,
                self.salida
            ],

            layout=widgets.Layout(
                align_items='center',
                gap='15px'
            )
        )

        display(interfaz)



class Principal:

    @staticmethod
    def main():

        ventana = VentanaPrincipal()

        ventana.inicio()



Principal.main()
```
