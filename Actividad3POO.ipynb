# <font color=#6130b0 size=6>**Ejercicio 8.2. Definicion de clases**</font>

Se requiere desarrollar un programa con interfaz gráfica de usuario que genere una ventana donde se solicite el ingreso de cinco notas obtenidas por un estudiante. El programa debe calcular y mostrar en la parte inferior de la ventana los siguientes datos:



*   El promedio de notas ingresadas.
*   La desviación estándar de las notas ingresadas.
*   La mayor nota obtenida.
*   La menor nota obtenida.




```python
class SistemaNotas:

    def __init__(self):


        self.titulo = widgets.HTML(
            value="""
            <h1 style='color:#6130b0; text-align:center;'>
                Sistema de Gestión de Notas
            </h1>

            <h3 style='text-align:center; color:gray;'>
                Programación Orientada a Objetos
            </h3>
            """
        )


        estilo = {'description_width': '80px'}

        layout_caja = widgets.Layout(
            width='250px',
            margin='10px auto'
        )


        self.nota1 = widgets.FloatText(
            description='Nota 1:',
            style=estilo,
            layout=layout_caja
        )

        self.nota2 = widgets.FloatText(
            description='Nota 2:',
            style=estilo,
            layout=layout_caja
        )

        self.nota3 = widgets.FloatText(
            description='Nota 3:',
            style=estilo,
            layout=layout_caja
        )

        self.nota4 = widgets.FloatText(
            description='Nota 4:',
            style=estilo,
            layout=layout_caja
        )

        self.nota5 = widgets.FloatText(
            description='Nota 5:',
            style=estilo,
            layout=layout_caja
        )


        self.boton = widgets.Button(
            description='Calcular Resultados',
            button_style='success',
            icon='check',
            layout=widgets.Layout(
                width='250px',
                height='45px',
                margin='20px auto'
            )
        )

        self.salida = widgets.Output(
            layout=widgets.Layout(
                display='flex',
                justify_content='center'
            )
        )


        self.boton.on_click(self.calcular_resultados)


    def obtener_notas(self):

        notas = [
            self.nota1.value,
            self.nota2.value,
            self.nota3.value,
            self.nota4.value,
            self.nota5.value
        ]

        return notas

# VALIDACIÓN DE NOTAS

    def validar_notas(self, notas):

        for nota in notas:

            if nota < 0 or nota > 5:
                return False

        return True

# CALCULO DE RESULTADOS

    def calcular_resultados(self, b):

        with self.salida:

            self.salida.clear_output()

            notas = self.obtener_notas()

            # VALIDACIÓN
            if not self.validar_notas(notas):

                display(HTML("""
                <h3 style='color:red; text-align:center;'>
                    Error: Las notas deben estar entre 0 y 5.
                </h3>
                """))

                return

            # CÁLCULOS
            promedio = sum(notas) / len(notas)
            desviacion = statistics.stdev(notas)
            mayor = max(notas)
            menor = min(notas)

            # RESULTADOS
            display(HTML(f"""
            <div style="
                background-color:#F5F5F5;
                padding:25px;
                border-radius:15px;
                border:2px solid #6130b0;
                width:450px;
                margin:auto;
                text-align:left;
                color:#6b6b6b;
            ">

            <h2 style='color:#6130b0; text-align:center;'>
                Resultados
            </h2>

            <p style='font-size:18px;'>
                📘 <b>Promedio:</b> {promedio:.2f}
            </p>

            <p style='font-size:18px;'>
                📊 <b>Desviación estándar:</b> {desviacion:.2f}
            </p>

            <p style='font-size:18px;'>
                🟢 <b>Mayor nota:</b> {mayor:.2f}
            </p>

            <p style='font-size:18px;'>
                🔴 <b>Menor nota:</b> {menor:.2f}
            </p>

            </div>
            """))



    def mostrar_interfaz(self):

        formulario = widgets.VBox(

            [
                self.titulo,
                self.nota1,
                self.nota2,
                self.nota3,
                self.nota4,
                self.nota5,
                self.boton,
                self.salida
            ],

            layout=widgets.Layout(
                align_items='center'
            )
        )

        display(formulario)



# CREAR OBJETO

sistema = SistemaNotas()

# INTERFAZ
sistema.mostrar_interfaz()
```

# <font color=#6130b0 size=6>**Ejercicio 8.3. Gestión de eventos**</font>
Se requiere desarrollar un programa con interfaz gráfica de usuario que permita calcular el volumen y superficie de varias figuras geométricas. Las figuras geométricas son el cilindro, la esfera y la pirámide.


*   Para el cilindro se solicitan su radio y altura (en centímetros).
*   Para la esfera, su radio (en centímetros).
*    Para la pirámide, su base, altura y apotema (en centímetros).


Una vez ingresados estos datos, el programa calcula el volumen y superficie de cada figura. Para desarrollar el programa se debe crear una jerarquía de clases para las diferentes figuras geométricas requeridas.

```python
# CLASE BASE

class FiguraGeometrica:

    def calcular_volumen(self):
        pass

    def calcular_superficie(self):
        pass


# =========================================================
# CLASE CILINDRO
# =========================================================

class Cilindro(FiguraGeometrica):

    def __init__(self, radio, altura):
        self.radio = radio
        self.altura = altura

    def calcular_volumen(self):
        return math.pi * (self.radio ** 2) * self.altura

    def calcular_superficie(self):
        return 2 * math.pi * self.radio * (self.radio + self.altura)


# CLASE ESFERA

class Esfera(FiguraGeometrica):

    def __init__(self, radio):
        self.radio = radio

    def calcular_volumen(self):
        return (4/3) * math.pi * (self.radio ** 3)

    def calcular_superficie(self):
        return 4 * math.pi * (self.radio ** 2)



# CLASE PIRÁMIDE

class Piramide(FiguraGeometrica):

    def __init__(self, base, altura, apotema):
        self.base = base
        self.altura = altura
        self.apotema = apotema

    def calcular_volumen(self):
        return (self.base ** 2 * self.altura) / 3

    def calcular_superficie(self):
        area_base = self.base ** 2
        area_lateral = 2 * self.base * self.apotema

        return area_base + area_lateral



# INTERFAZ GRÁFICA

titulo = widgets.HTML("""
<h1 style='color:#6130b0; text-align:center;'>
Sistema de Figuras Geométricas
</h1>

<h3 style='text-align:center; color:gray;'>
Programación Orientada a Objetos
</h3>
""")

estilo = {'description_width': '120px'}

layout_caja = widgets.Layout(
    width='300px',
    margin='8px auto'
)


# CILINDRO

subtitulo_cilindro = widgets.HTML("""
<h3 style='color:#6130b0;'>Cilindro</h3>
""")

radio_cilindro = widgets.FloatText(
    description='Radio:',
    style=estilo,
    layout=layout_caja
)

altura_cilindro = widgets.FloatText(
    description='Altura:',
    style=estilo,
    layout=layout_caja
)


# ESFERA

subtitulo_esfera = widgets.HTML("""
<h3 style='color:#6130b0;'>Esfera</h3>
""")

radio_esfera = widgets.FloatText(
    description='Radio:',
    style=estilo,
    layout=layout_caja
)


# PIRÁMIDE

subtitulo_piramide = widgets.HTML("""
<h3 style='color:#6130b0;'>Pirámide</h3>
""")

base_piramide = widgets.FloatText(
    description='Base:',
    style=estilo,
    layout=layout_caja
)

altura_piramide = widgets.FloatText(
    description='Altura:',
    style=estilo,
    layout=layout_caja
)

apotema_piramide = widgets.FloatText(
    description='Apotema:',
    style=estilo,
    layout=layout_caja
)



boton = widgets.Button(
    description='Calcular',
    button_style='success',
    icon='calculator',
    layout=widgets.Layout(
        width='300px',
        height='45px',
        margin='20px auto'
    )
)


salida = widgets.Output(
    layout=widgets.Layout(
        display='flex',
        justify_content='center'
    )
)


def calcular_figuras(b):

    with salida:

        salida.clear_output()

        valores = [
            radio_cilindro.value,
            altura_cilindro.value,
            radio_esfera.value,
            base_piramide.value,
            altura_piramide.value,
            apotema_piramide.value
        ]

        for valor in valores:

            if valor <= 0:

                display(HTML("""
                <h3 style='color:red; text-align:center;'>
                    Error: Todos los valores deben ser mayores que cero.
                </h3>
                """))

                return

        # FIGURAS
        cilindro = Cilindro(
            radio_cilindro.value,
            altura_cilindro.value
        )

        esfera = Esfera(
            radio_esfera.value
        )

        piramide = Piramide(
            base_piramide.value,
            altura_piramide.value,
            apotema_piramide.value
        )

        # RESULTADOS
        display(HTML(f"""
        <div style="
            background-color:#F5F5F5;
            padding:25px;
            border-radius:15px;
            border:2px solid #6130b0;
            width:550px;
            margin:auto;
            color:#6b6b6b;
            text-align:left;
        ">

        <h2 style='color:#6130b0; text-align:center;'>
            Resultados
        </h2>

        <hr>

        <h3 style='color:#6130b0;'>Cilindro</h3>

        <p><b>Volumen:</b> {cilindro.calcular_volumen():.2f} cm³</p>
        <p><b>Superficie:</b> {cilindro.calcular_superficie():.2f} cm²</p>

        <hr>

        <h3 style='color:#6130b0;'>Esfera</h3>

        <p><b>Volumen:</b> {esfera.calcular_volumen():.2f} cm³</p>
        <p><b>Superficie:</b> {esfera.calcular_superficie():.2f} cm²</p>

        <hr>

        <h3 style='color:#6130b0;'>Pirámide</h3>

        <p><b>Volumen:</b> {piramide.calcular_volumen():.2f} cm³</p>
        <p><b>Superficie:</b> {piramide.calcular_superficie():.2f} cm²</p>

        </div>
        """))


boton.on_click(calcular_figuras)


formulario = widgets.VBox(

    [
        titulo,

        subtitulo_cilindro,
        radio_cilindro,
        altura_cilindro,

        subtitulo_esfera,
        radio_esfera,

        subtitulo_piramide,
        base_piramide,
        altura_piramide,
        apotema_piramide,

        boton,
        salida
    ],

    layout=widgets.Layout(
        align_items='center'
    )
)


# INTERFAZ


display(formulario)
```
